# Stochastic Gradient Descent

## SGD Pseudocode
Stochastic Gradient Descent (SGD) is an optimization algorithm commonly used in machine learning and deep learning to train models. Here's the pseudocode for a basic version of SGD:

```
Input:
  - Learning rate: α (a small positive number)
  - Number of epochs: N (the number of times to iterate over the entire dataset)
  - Initial model parameters: θ (randomly or with some initialization)

for epoch in 1 to N do:
  Shuffle the training data
  for each training example (x, y) in the dataset do:
    Compute the gradient of the loss function with respect to the parameters:
      ∇θ Loss(θ, x, y)
    Update the parameters using the gradient and the learning rate:
      θ = θ - α * ∇θ Loss(θ, x, y)
  end for
end for
```

In this pseudocode:

- `Learning rate (α)` determines the step size in the parameter space for each update.
- `Number of epochs (N)` specifies how many times the algorithm iterates over the entire training dataset.
- `θ` represents the model parameters that we are trying to optimize.
- The inner loop processes one training example at a time, computing the gradient of the loss function with respect to the parameters for that example and then updating the parameters accordingly.
- Shuffling the training data at the beginning of each epoch helps introduce randomness into the training process, which can improve convergence.

It's important to note that there are variations of SGD, such as Mini-batch SGD and Momentum SGD, which can be used to improve convergence speed and stability. Additionally, the choice of learning rate and other hyperparameters can significantly impact the performance of the algorithm in practice.
